```python
# raw_text is a LIST with one big string
text = raw_text[0]          # take the big text

paragraphs = [p.strip() for p in text.split("\n\n") if p.strip()]

print(len(paragraphs))
```

```python
for i, p in enumerate(paragraphs, 1):
    print(f"\nParagraph {i}")
    print(p)
```

```python
from sentence_transformers import SentenceTransformer, util
import torch


embed_model = SentenceTransformer("sentence-transformers/paraphrase-multilingual-MiniLM-L12-v2")

embeddings = embed_model.encode(paragraphs, convert_to_tensor=True)

groups = []
current_group = [paragraphs[0]]
current_emb = embeddings[0]

THRESHOLD = 0.85

for i in range(1, len(paragraphs)):
    sim = util.cos_sim(embeddings[i], current_emb).item()

    if sim >= THRESHOLD:
        current_group.append(paragraphs[i])
        current_emb = torch.mean(
            torch.stack([current_emb, embeddings[i]]), dim=0
        )
    else:
        groups.append(current_group)
        current_group = [paragraphs[i]]
        current_emb = embeddings[i]

groups.append(current_group)

print("Total news articles:", len(groups))
```

```python
for i, g in enumerate(groups, 1):
    print(f"\n📰 News {i}")
    for p in g:
        print("-", p[:80], "...")
```

```python
from transformers import AutoTokenizer, AutoModelForSeq2SeqLM

# PUBLIC & WORKING model
tokenizer = AutoTokenizer.from_pretrained("ai4bharat/indicbart")
headline_model = AutoModelForSeq2SeqLM.from_pretrained("ai4bharat/indicbart")
```

```python
def is_headline(text):
    text = text.strip()

    # reject symbol-only / garbage lines
    if len(set(text)) <= 3:
        return False

    return (
        len(text.split()) <= 8 and
        "。" not in text and
        "." not in text
    )
```

```python
def get_first_clean_para(group):
    for p in group:
        p = p.strip()
        # skip garbage / symbol-only lines
        if len(p) > 15 and len(set(p)) > 5:
            return p
    return group[0]
```

```python
def is_mostly_text(text, min_ratio=0.6):
    # count Tamil letters
    tamil_chars = sum(1 for c in text if '\u0B80' <= c <= '\u0BFF')
    return tamil_chars / max(len(text), 1) >= min_ratio
```

```python
def generate_headline_safe(group):
    # 1) Use existing headline if present
    first_para = group[0].strip()
    if is_headline(first_para):
        return first_para

    # 2) Otherwise generate using IndicBART
    text = " ".join(group)

    inputs = tokenizer(
    text,
    return_tensors="pt",
    truncation=True,
    max_length=512
    )

    # REMOVE token_type_ids (IndicBART doesn't use it)
    inputs.pop("token_type_ids", None)


    outputs = headline_model.generate(
        **inputs,
        max_new_tokens=15,
        do_sample=False
    )

    headline = tokenizer.decode(outputs[0], skip_special_tokens=True).strip()
    headline = headline.replace("[CLS]", "").replace("[SEP]", "").strip()


    # bad output detection
    if (
      not headline or
      "<extra_id" in headline or
      "[CLS]" in headline or
      headline in ["-", "–"] or
      len(set(headline)) <= 2 or
      not is_mostly_text(headline)
          ):
      return get_first_clean_para(group)


    # enforce headline length (max 8 words)
    # enforce headline length (newspaper style)
   # FINAL enforcement – always applied
    words = headline.split()

    # too long → cut
    if len(words) > 8:
        headline = " ".join(words[:8])

    # too short → use clean paragraph
    if len(words) < 4:
        headline = get_first_clean_para(group).split("।")[0]

    return headline


```

```python
results = []

for group in groups:
    title = generate_headline_safe(group)
    results.append({
        "headline": title,
        "content": group
    })
```

```python
for i, item in enumerate(results, 1):
    print(f"\n📰 News {i}")
    print("Headline:", item["headline"])
    print("Paragraphs:", len(item["content"]))
```
