```python
import glob
import pyarrow.parquet as pq
from datasets import Dataset, concatenate_datasets


from transformers import (
    AutoTokenizer,
    AutoModelForCausalLM,
    Trainer,
    TrainingArguments,
)

MODEL_NAME = "microsoft/phi-2"
PARQUET_DIR = r"D:\final Year project\Research_backend\DataSet"
TEXT_COLUMN = "text"   # CHANGE if your column name is different

# Load model
tokenizer = AutoTokenizer.from_pretrained(MODEL_NAME)
model = AutoModelForCausalLM.from_pretrained(MODEL_NAME)

datasets = []

# Loop all parquet files
for file in sorted(glob.glob(f"{PARQUET_DIR}/ta_part_*.parquet")):
    pf = pq.ParquetFile(file)
    batch = next(pf.iter_batches(batch_size=5000))  # safe chunk
    df = batch.to_pandas()
    datasets.append(Dataset.from_pandas(df[[TEXT_COLUMN]]))

# Merge all datasets
dataset = concatenate_datasets(datasets)

# Tokenize
def tokenize_fn(examples):
    return tokenizer(
        examples[TEXT_COLUMN],
        truncation=True,
        padding="max_length",
        max_length=512,
    )

dataset = dataset.map(tokenize_fn, batched=True, remove_columns=[TEXT_COLUMN])

# Training settings (CPU safe)
training_args = TrainingArguments(
    output_dir="./phi2_tamil_ft",
    per_device_train_batch_size=1,
    gradient_accumulation_steps=4,
    num_train_epochs=2,
    logging_steps=50,
    save_strategy="epoch",
    fp16=False,
)

trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset=dataset,
)

trainer.train()
```

```python
import torch
from transformers import AutoTokenizer, AutoModelForCausalLM

model_name = "Qwen/Qwen2.5-3B-Instruct"

# Load tokenizer & model (use already downloaded cache)
tokenizer = AutoTokenizer.from_pretrained(
    model_name,
    trust_remote_code=True
)

model = AutoModelForCausalLM.from_pretrained(
    model_name,
    trust_remote_code=True,
    torch_dtype=torch.float16 if torch.cuda.is_available() else torch.float32,
    device_map="auto"
)

tokenizer.pad_token = tokenizer.eos_token

# -------- TEST PROMPT --------
prompt = """### Instruction:
தமிழில் ஒரு எளிய வாக்கியம் எழுதவும்.

### Response:
"""

inputs = tokenizer(prompt, return_tensors="pt").to(model.device)

with torch.no_grad():
    outputs = model.generate(
        **inputs,
        max_new_tokens=50,
        do_sample=True,
        temperature=0.7
    )

print(tokenizer.decode(outputs[0], skip_special_tokens=True))
```

```python
import torch
import pandas as pd
from datasets import Dataset
from transformers import (
    AutoTokenizer,
    AutoModelForCausalLM,
    TrainingArguments,
    Trainer
)
from peft import LoraConfig, get_peft_model, TaskType

# ===============================
# CONFIG
# ===============================
MODEL_NAME = "Qwen/Qwen2.5-3B-Instruct"
CSV_PATH = r"D:\Downloads\tamil.csv"
OUTPUT_DIR = "./qwen25-tamil-lora"

# ===============================
# LOAD DATASET
# ===============================
df = pd.read_csv(CSV_PATH, encoding="utf-8")
df = df[["full_news", "semantic_title"]]

dataset = Dataset.from_pandas(df)

def build_prompt(example):
    text = f"""### Instruction:
கீழே கொடுக்கப்பட்ட செய்திக்கான பொருத்தமான தலைப்பை தமிழில் உருவாக்கவும்.

### News:
{example['full_news']}

### Response:
{example['semantic_title']}"""
    return {"text": text}

dataset = dataset.map(build_prompt)

# ===============================
# TOKENIZER
# ===============================
tokenizer = AutoTokenizer.from_pretrained(
    MODEL_NAME,
    trust_remote_code=True
)
tokenizer.pad_token = tokenizer.eos_token

def tokenize(example):
    tokens = tokenizer(
        example["text"],
        truncation=True,
        max_length=512,
        padding="max_length"
    )
    tokens["labels"] = tokens["input_ids"].copy()
    return tokens

tokenized_dataset = dataset.map(
    tokenize,
    batched=True,
    remove_columns=dataset.column_names
)

# ===============================
# LOAD MODEL
# ===============================

model = AutoModelForCausalLM.from_pretrained(
    MODEL_NAME,
    trust_remote_code=True,
    torch_dtype=torch.float32,   
    device_map=None              
)

# ===============================
# LORA CONFIG
# ===============================
lora_config = LoraConfig(
    task_type=TaskType.CAUSAL_LM,
    r=8,
    lora_alpha=16,
    lora_dropout=0.05,
    target_modules=["q_proj", "k_proj", "v_proj", "o_proj"]
)

model = get_peft_model(model, lora_config)
model.print_trainable_parameters()

# ===============================
# TRAINING ARGS
# ===============================
training_args = TrainingArguments(
    output_dir=OUTPUT_DIR,
    per_device_train_batch_size=1,
    gradient_accumulation_steps=8,
    num_train_epochs=3,
    learning_rate=2e-4,
    fp16=True,
    logging_steps=50,
    save_steps=1000,
    save_total_limit=2,
    report_to="none",
    no_cuda=True
)

# ===============================
# TRAIN
# ===============================
trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset=tokenized_dataset,
    tokenizer=tokenizer
)

trainer.train()

# ===============================
# SAVE LORA MODEL
# ===============================
model.save_pretrained(OUTPUT_DIR)
tokenizer.save_pretrained(OUTPUT_DIR)

print("LoRA fine-tuning completed successfully")
```
